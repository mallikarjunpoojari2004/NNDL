```python
# Full corrected pipeline — paste and run this single cell
import os, sys
import pandas as pd
import numpy as np
import difflib
import matplotlib.pyplot as plt

# Config
PATH = "colors.txt"   # change to "colors.csv" if needed
MAXLEN = 30
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

# 1) Read file robustly
if not os.path.exists(PATH):
    raise FileNotFoundError(f"{PATH} not found in working dir: {os.getcwd()}")

# Try tab first (your sample looks tab-separated)
try:
    raw = pd.read_csv(PATH, sep="\t", encoding="utf-8-sig", header=0)
except Exception:
    raw = pd.read_csv(PATH, sep=None, engine="python", encoding="utf-8-sig", header=0)

# Keep first 4 columns only if file contains extras
if raw.shape[1] > 4:
    raw = raw.iloc[:, :4]

# Ensure columns exist and name them
raw.columns = ["name", "red", "green", "blue"]

# Strip and keep original
raw["name"] = raw["name"].astype(str).str.strip()

# Convert RGB -> numeric, invalid -> NaN
for c in ["red","green","blue"]:
    raw[c] = pd.to_numeric(raw[c], errors="coerce")

# Show & drop bad rows (print them for inspection)
bad_rows = raw[raw[["red","green","blue"]].isnull().any(axis=1)]
print("Number of rows with invalid/missing RGB:", len(bad_rows))
if len(bad_rows) > 0:
    print("\n=== Sample bad rows (inspect) ===")
    print(bad_rows.to_string(index=False))
    print("\nDropping these bad rows and continuing...")
    raw = raw.dropna(subset=["red","green","blue"])

# Drop empty names
raw = raw[raw["name"].str.strip() != ""].copy()

# Clip and convert to int
raw[["red","green","blue"]] = raw[["red","green","blue"]].clip(0,255).round().astype(int)

print("\nRaw data loaded. Rows:", raw.shape[0])
print(raw.head(12).to_string(index=False))

# 2) Create normalized name column and group properly (avoid the earlier warning)
raw['name_l'] = raw['name'].str.strip().str.lower()

# Group by the normalized name and average RGBs
df_grouped = raw.groupby('name_l', as_index=False).agg({
    'red': 'mean',
    'green': 'mean',
    'blue': 'mean'
})

# Round and convert to int
df_grouped[['red','green','blue']] = df_grouped[['red','green','blue']].round().astype(int)

# Keep final DF with 'name' column (lowercased)
df_grouped = df_grouped.rename(columns={'name_l':'name'})[['name','red','green','blue']]
df_grouped['name'] = df_grouped['name'].astype(str).str.strip()

print("\nUnique names after grouping:", df_grouped.shape[0])
print(df_grouped.head(12).to_string(index=False))

# 3) Tokenize char-level and prepare sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

names = df_grouped['name'].tolist()
tokenizer = Tokenizer(char_level=True, filters=None, lower=True, oov_token=None)
tokenizer.fit_on_texts(names)
vocab_size = len(tokenizer.word_index) + 1
print("\nVocab size (chars):", vocab_size)

seqs = tokenizer.texts_to_sequences(names)
X = pad_sequences(seqs, padding="post", maxlen=MAXLEN).astype("int32")
y = df_grouped[["red","green","blue"]].values.astype("float32") / 255.0

print("X shape:", X.shape, "y shape:", y.shape)
print("y min/max per channel:", np.nanmin(y, axis=0), np.nanmax(y, axis=0))

# 4) Build safe model (Embedding + BiLSTM) with MAE and gradient clipping
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split

EMBED_DIM = 64
LSTM_UNITS = 128

model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=EMBED_DIM, input_length=MAXLEN, mask_zero=True),
    Bidirectional(LSTM(LSTM_UNITS, return_sequences=False)),
    Dropout(0.3),
    Dense(128, activation="relu"),
    Dropout(0.2),
    Dense(3, activation="sigmoid")
])

opt = Adam(learning_rate=5e-4, clipnorm=1.0)
model.compile(optimizer=opt, loss="mae", metrics=["mae"])
model.summary()

# 5) Train with validation split and callbacks
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.15, random_state=RANDOM_SEED, shuffle=True)

es = EarlyStopping(monitor="val_loss", patience=12, restore_best_weights=True)
rlr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=5, min_lr=1e-6, verbose=1)
chk = ModelCheckpoint("best_color_model.h5", monitor="val_loss", save_best_only=True, verbose=0)

history = model.fit(X_train, y_train, validation_data=(X_val, y_val),
                    epochs=200, batch_size=16, callbacks=[es, rlr, chk], verbose=2)

# 6) Training diagnostics
plt.figure(figsize=(6,4))
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='val')
plt.xlabel('epoch'); plt.ylabel('MAE'); plt.legend(); plt.title('Training MAE'); plt.show()

train_mae = model.evaluate(X_train, y_train, verbose=0)[0]
val_mae = model.evaluate(X_val, y_val, verbose=0)[0]
print(f"\nTrain MAE (0-1): {train_mae:.4f}, Val MAE (0-1): {val_mae:.4f}")
print(f"Train MAE (0-255): {train_mae*255:.2f}, Val MAE (0-255): {val_mae*255:.2f}")

# 7) Diagnostics: worst predictions
X_all = pad_sequences(tokenizer.texts_to_sequences(df_grouped['name'].tolist()), padding='post', maxlen=MAXLEN)
preds = model.predict(X_all, verbose=0)
preds_rgb = (np.clip(preds,0,1) * 255.0).round().astype(int)
true_rgb = df_grouped[['red','green','blue']].values.astype(int)
mae_per_sample = np.mean(np.abs(preds_rgb - true_rgb), axis=1)

df_diag = df_grouped.copy()
df_diag['pred_r'] = preds_rgb[:,0]; df_diag['pred_g'] = preds_rgb[:,1]; df_diag['pred_b'] = preds_rgb[:,2]
df_diag['mae'] = mae_per_sample
df_diag = df_diag.sort_values('mae', ascending=False)

print("\nTop 15 worst predictions (name, true_rgb, pred_rgb, mae):")
print(df_diag[['name','red','green','blue','pred_r','pred_g','pred_b','mae']].head(15).to_string(index=False))

print("\nOverall MAE per channel (0-255):", np.mean(np.abs(preds_rgb - true_rgb), axis=0))
print("Overall MAE (0-255):", np.mean(mae_per_sample))

# 8) Hybrid prediction & display (prefer exact dataset color for very close names)
name_to_rgb = {n: tuple(int(x) for x in row) for n,row in zip(df_grouped['name'], df_grouped[['red','green','blue']].values)}
all_names = df_grouped['name'].tolist()

def predict_model_rgb(name, debug=False):
    seq = tokenizer.texts_to_sequences([name.lower()])
    seq = pad_sequences(seq, padding="post", maxlen=MAXLEN)
    pred = model.predict(seq, verbose=0)[0]
    if debug:
        print("raw pred (0-1):", pred)
    if not np.all(np.isfinite(pred)):
        return None
    pred_rgb = tuple(int(x) for x in (np.clip(pred,0,1)*255.0).round())
    return pred_rgb

def predict_hybrid(name, cutoff=0.90):
    match = difflib.get_close_matches(name.lower(), all_names, n=1, cutoff=0.0)
    if match:
        ratio = difflib.SequenceMatcher(None, name.lower(), match[0]).ratio()
        if ratio >= cutoff:
            return name_to_rgb[match[0]], match[0], 'baseline'
    pred = predict_model_rgb(name)
    if pred is None:
        if match:
            return name_to_rgb[match[0]], match[0], 'baseline'
        return None, None, 'none'
    return pred, (match[0] if match else None), 'model'

def show_color(qname, cutoff=0.90):
    rgb, matched, source = predict_hybrid(qname, cutoff=cutoff)
    if rgb is None:
        print("No prediction available for", qname)
        return
    print(f"Input: {qname}  | source: {source}  | matched: {matched}  | rgb: {rgb}")
    patch = np.zeros((120,120,3), dtype=np.uint8)
    patch[:,:,0] = rgb[0]; patch[:,:,1] = rgb[1]; patch[:,:,2] = rgb[2]
    plt.imshow(patch)
    plt.title(f"{qname} — {rgb} ({source})")
    plt.axis("off")
    plt.show()

# 9) Examples
show_color("yellow")
show_color("gren")
show_color("tangerine")
show_color("black")

print("\nDone. If you want me to raise the hybrid cutoff (e.g. to 0.95) so known names always return exact dataset RGB, say 'raise cutoff'.")
```
